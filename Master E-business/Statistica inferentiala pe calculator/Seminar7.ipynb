Un eșantion de 50 de elemente dintr-o populație cu σ=6 are media 32. Construiți intervalele de încredere ale mediei de 90%, 95%, și 99%.

```python
n=50
sigma=6
x_barat=32
alfa=0.1
```

```python
# z, sigma
eroarea_standard=sigma/np.sqrt(n)
z_alfa_pe_2=norm.ppf(1-alfa/2)
```

```python
z_alfa_pe_2
```

```python
jos=x_barat - z_alfa_pe_2*eroarea_standard
sus=x_barat + z_alfa_pe_2*eroarea_standard
```

```python
jos,sus
```

```python
Alfa=[0.1,0.05, 0.01]
for alfa in Alfa:
    z_alfa_pe_2 = norm.ppf(1-alfa/2)
    Jos = x_barat - z_alfa_pe_2 * eroarea_standard
    Sus = x_barat + z_alfa_pe_2 * eroarea_standard
    print(f'alfa={alfa}, [{Jos},{Sus}]')
```

>3.	Verificați de câte ori media populației apare în intervalul calculat.
>4.	Pentru un eșantion (oarecare) reprezentați grafic distribuția de eșantionare a mediei, distribuția corespunzătoare mediei eșantionului si intervalul de încredere de 95% (pe același grafic)

o populație normal distribuită cu media 5 și σ=2;

```python
#1
import numpy as np
from scipy.stats import norm, uniform
import matplotlib.pyplot as plt

n_values = [20, 30, 50, 100]
pop_mean = 5
pop_std = 2
pop_low, pop_high = 12, 20

samples_normal = {n: np.random.normal(pop_mean, pop_std, size=(10000, n)) for n in n_values}
samples_uniform = {n: np.random.uniform(pop_low, pop_high, size=(10000, n)) for n in n_values}
```

```python
samples_normal
```

```python
samples_uniform
```

```python
#2 + 3
confidence_levels = [0.9, 0.95, 0.99]
for n in n_values:
    sample_means = np.mean(samples_normal[n], axis=1)
    for level in confidence_levels:
        lower, upper = norm.interval(level, loc=np.mean(sample_means), scale=np.std(sample_means)/np.sqrt(n))
        print(f'Intervalul de încredere {level*100}% pentru o populație normal distribuită cu n={n} este: [{lower:.2f}, {upper:.2f}]')
        print(f'Media populației apare in acest interval in {np.mean(lower <= pop_mean <= upper):.2%} din cazuri')
```

```python
#4
n = 20
sample_means = np.mean(samples_normal[n], axis=1)
plt.hist(sample_means, bins=50, density=True, label='Distribuția de eșantionare a mediei')

x = np.linspace(np.min(sample_means), np.max(sample_means), 100)
plt.plot(x, norm.pdf(x, np.mean(sample_means), np.std(sample_means)/np.sqrt(n)),'r',lw=2)
plt.show()
```
